```python
#from ultralytics import YOLO
#import numpy as np

# Åadowanie modelu
#model3 = YOLO("/home/wnadworska/programming/kakao_v2/runs/detect/train22/weights/best.pt") #wytrenowane najlepsze wagi

# WybÃ³r zdjÄ™Ä‡ do predykcji
import os
from pathlib import Path

#ÅšcieÅ¼ka do zdjÄ™Ä‡ (folder z samymi zdjÄ™ciami, bez labels)
folder_path = "/home/wnadworska/programming/kakao_v2/test/images/"

#Wybranie plikÃ³w .jpg
image_paths = [str(p) for p in Path(folder_path).glob("*.jpg")]

# Predykcja na zdjÄ™ciach
results = best_model2(image_paths)  #podana lista obiektÃ³w

# Lista z informacjami o BB
boxes = []
for result in results:
    boxes.append(result.boxes)

# Lista zliczeÅ„ BB na kaÅ¼dym ze zdjÄ™Ä‡
count_list = []
from collections import Counter, defaultdict

for i in range(len(results)):
    bb = boxes[i].cls.tolist()
    count_bb = Counter(bb) #zliczenie BB na zdjÄ™ciu
    count_list.append(count_bb) #dorzucenie do listy zliczeÅ„

# SÅ‚ownik z sumowaniem BB kaÅ¼dej z klas
summary = defaultdict(int)

for i in range(len(count_list)):
     #WyciÄ…gniÄ™cie elementu z listy zliczeÅ„
     count_dict = count_list[i].items() #np. Counter({0.0: 3})
     for key, value in count_dict:
          key_str = f"{key:.1f}"[:3]  #wyciÄ…gniÄ™cie 3 znakÃ³w na lewo od :
          summary[key_str] += value  #sumowanie wg numeru klasy

# Wyniki zliczenia klas - {klasa: liczba BBoxÃ³w}
print(summary)
print("Liczba BB klasy 'cocoa-pod-mature-unriped': ", summary['0.0'])
print("Liczba BB klasy 'cocoa-tree': ", summary['1.0'])
print("Liczba BB klasy 'cocoa-pod-immature': ", summary['2.0'])
print("Liczba BB klasy 'cocoa-pod-riped': ", summary['3.0'])
print("Liczba BB klasy 'cocoa-pod-spoilt': ", summary['4.0'])
```
